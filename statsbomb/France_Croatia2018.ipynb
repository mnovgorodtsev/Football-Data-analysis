```python
match_events = sb.events(match_id=8658)
```

```python
xg_df=match_events[match_events['shot_statsbomb_xg'].notnull()]
xg_df=xg_df[['team','type','minute','location','shot_statsbomb_xg','type']]
Cro_xg=xg_df[xg_df['team']=='Croatia']
Fra_xg=xg_df[xg_df['team']=='France']

Cro_xg[['x_shot','y_shot']]=pd.DataFrame(Cro_xg.location.tolist(), index=Cro_xg.index)
Fra_xg[['x_shot','y_shot']]=pd.DataFrame(Fra_xg.location.tolist(), index=Fra_xg.index)
```

```python
from mplsoccer import VerticalPitch
from matplotlib.patches import Patch
import matplotlib.pyplot as plt

pitch = VerticalPitch(pitch_type='statsbomb',half=True)
fig, ax = pitch.draw(figsize=(9,7))


for index, row in Fra_xg.iterrows():
    fra_scatter=pitch.scatter(x=Fra_xg['x_shot'], y=Fra_xg['y_shot'],ax=ax,c='blue',s=Fra_xg['shot_statsbomb_xg']*1000, label='France',edgecolors='black',alpha=0.22, marker='h')

for index, row in Cro_xg.iterrows():
    cro_scatter=pitch.scatter(x=Cro_xg['x_shot'], y=Cro_xg['y_shot'],ax=ax,c='red',s=Cro_xg['shot_statsbomb_xg']*1000, label='Croatia',edgecolors='black',alpha=0.15, marker='h')


    
plt.title('xG France-Croatia, World Cup Final 2018')
legend_elements = [Patch(color='blue', label='France'), Patch(color='red', label='Croatia')]
plt.legend(handles=legend_elements, loc='upper right', fontsize="15")

ax.set_facecolor("#538032")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
from scipy.interpolate import interp1d 

#x = np.linspace(20, 1, 90)
x=xg_df['minute']
france_xg=xg_df[xg_df['team']=='France'][['shot_statsbomb_xg','minute']]
croatia_xg=xg_df[xg_df['team']=='Croatia'][['shot_statsbomb_xg','minute']]
france_xg = france_xg.groupby('minute', as_index=False).sum()
croatia_xg = croatia_xg.groupby('minute', as_index=False).sum()

fig, ax = plt.subplots(figsize=(6, 4))



x_fra=france_xg['minute']
y_fra=france_xg['shot_statsbomb_xg']
x_cro=croatia_xg['minute']
y_cro=croatia_xg['shot_statsbomb_xg']

ax.plot(x_fra, france_xg['shot_statsbomb_xg'], c='blue')
ax.plot(x_cro, croatia_xg['shot_statsbomb_xg'], c='red')
ax.set_xlim(19, 90)
plt.title('WC 2018 Final, xG graph')
ax.set_facecolor('#eeeeee')
ax.set_xlabel('Minute')
ax.set_ylabel('xG')
legend_elements = [Patch(color='blue', label='France'), Patch(color='red', label='Croatia')]
plt.legend(handles=legend_elements, loc='upper right', fontsize="10")
plt.show()
```
